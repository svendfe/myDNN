```python
import os

os.environ["KERAS_BACKEND"] = "tensorflow"

import tensorflow as tf
import pandas as pd
import keras
from keras.utils import FeatureSpace

import tensorflow_decision_forests as tfdf
import ydf
```

```python
input_file_path = "FootballPlayerPreparedCleanAttributes.csv"
output_file_path = "FootballPlayerOneHotEncodedClasses.csv"
input_dataframe = pd.read_csv(input_file_path)
output_dataframe = pd.read_csv(output_file_path)
```

```python
seed = 42

train_x = input_dataframe.sample(frac = 0.8, random_state=seed)
test_x = input_dataframe.drop(train_x.index).sample(frac = 0.5, random_state=seed)
validate_x = input_dataframe.drop(train_x.index).drop(test_x.index)

train_y = output_dataframe.loc[train_x.index]
test_y = output_dataframe.loc[test_x.index]
validate_y = output_dataframe.loc[validate_x.index]
```

```python
train_y_single = train_y.idxmax(axis=1)
train_data = train_x.copy()
train_data["label"] = train_y_single
```

```python
train_data.describe()
```

```python
test_y_single = test_y.idxmax(axis=1)
test_data = test_x.copy()
test_data["label"] = test_y_single
```

```python
#ydf
learner = ydf.RandomForestLearner(label="label")
model = learner.train(train_data)
predictions = model.predict(test_data)
evaluation = model.evaluate(test_data)
analysis = model.analyze(test_data)
```

```python
evaluation
```

```python
#ydf
learner = ydf.GradientBoostedTreesLearner(label="label")
model = learner.train(train_data)
predictions = model.predict(test_data)
evaluation = model.evaluate(test_data)
analysis = model.analyze(test_data)
```
